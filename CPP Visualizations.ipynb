## disp_imp

```python
# Prepare dataset-specific data
datasets = ['Bank', 'COMPAS (Race)', 'COMPAS (Sex)', 'German (Age)', 'German (Sex)']
dfs = [fair_bank, fair_compas_race, fair_compas_sex, fair_german_age, fair_german_sex]

# Extract Disparate Impact values
orig_values = []
cpp_values = []

for df in dfs:
    filtered_df = df.loc[df['Metric'].str.contains('disp_imp')]
    orig_values.append(filtered_df[filtered_df['Method'] == 'orig']['Mean'].values[0])
    cpp_values.append(filtered_df[filtered_df['Method'] == 'cpp']['Mean'].values[0])

# Plotting
x = np.arange(len(datasets))  # x locations for the groups
width = 0.35  # width of the bars

fig, ax = plt.subplots(figsize=(10, 6))

# Dark blue and dark red
rects1 = ax.bar(x - width/2, orig_values, width, label='Orig', color='#1f77b4', edgecolor='black')  # Dark blue
rects2 = ax.bar(x + width/2, cpp_values, width, label='CPP', color='#d62728', edgecolor='black')  # Dark red

# Add titles and labels
ax.set_xlabel('Datasets', fontsize=12)
ax.set_ylabel('Disparate Impact', fontsize=12)
ax.set_title('Fairness Metric', fontsize=14, weight='bold')
ax.set_xticks(x)
ax.set_xticklabels(datasets, rotation=45, fontsize=10)
ax.legend()

# Finalize layout
plt.tight_layout()
plt.show()
```

```python
cpp_bank = pd.read_csv("./results/cpp_mia2_bank_train_test_accuracies_2024-12-30.csv")

cpp_compas_race = pd.read_csv("./results/cpp_mia2_compas_race_train_test_accuracies_2024-12-30.csv")

cpp_compas_sex = pd.read_csv("./results/cpp_mia2_compas_sex_train_test_accuracies_2024-12-30.csv")

cpp_german_age = pd.read_csv("./results/cpp_mia2_german_age_train_test_accuracies_2024-12-30.csv")

cpp_german_sex = pd.read_csv("./results/cpp_mia2_german_sex_train_test_accuracies_2024-12-30.csv")

cpp_law_school_gender = pd.read_csv("./results/cpp_mia2_law_gender_aif_train_test_accuracies_2024-12-30.csv")

cpp_law_school_race = pd.read_csv("./results/cpp_mia2_law_race_aif_train_test_accuracies_2024-12-30.csv")

cpp_law_sex = pd.read_csv("./results/cpp_mia2_law_sex_train_test_accuracies_2024-12-30.csv")

cpp_law_race = pd.read_csv("./results/cpp_mia2_law_race_train_test_accuracies_2024-12-30.csv")

cpp_meps19 = pd.read_csv("./results/cpp_mia2_meps19_train_test_accuracies_2024-12-30.csv")

def find_max_gap_and_plot(datasets, names):
    dimensions = []
    values_orig = []
    values_cpp = []

    for data, name in zip(datasets, names):
        # Filter rows that correspond to subpopulation test accuracies
        test_subpopulations = data[data['Unnamed: 0'].str.contains("accuracy_test_")].copy()

        # Calculate the gaps between orig_acc_mean and cpp_acc_mean where orig > cpp
        test_subpopulations['gap'] = (test_subpopulations['orig_acc_mean'] -
                                      test_subpopulations['cpp_acc_mean'])
        valid_gaps = test_subpopulations[test_subpopulations['gap'] > 0]

        # Find the index of the maximum gap
        max_gap_index = valid_gaps['gap'].idxmax()

        if pd.notna(max_gap_index):
            max_gap_row = valid_gaps.loc[max_gap_index]
            dimensions.append(name)
            values_orig.append(max_gap_row['orig_acc_mean'])
            values_cpp.append(max_gap_row['cpp_acc_mean'])

    # Convert to numpy arrays for radial plot
    values_orig = np.array(values_orig)
    values_cpp = np.array(values_cpp)

    # Add the first value to close the radial plot
    values_orig = np.append(values_orig, values_orig[0])
    values_cpp = np.append(values_cpp, values_cpp[0])

    # Define the angles for the radial plot
    angles = np.linspace(0, 2 * np.pi, len(dimensions) + 1, endpoint=True)

    # Create the radial plot
    fig, ax = plt.subplots(figsize=(8, 8), subplot_kw={"projection": "polar"})

    # Place 0 (angle=0) at the top (N) and move clockwise
    ax.set_theta_zero_location("N")
    ax.set_theta_direction(-1)

    # Plot the data
    ax.plot(angles, values_orig, label="Original", linewidth=2)
    ax.fill(angles, values_orig, alpha=0.25)
    ax.plot(angles, values_cpp, label="CPP", linewidth=2)
    ax.fill(angles, values_cpp, alpha=0.25)

    # Determine a suitable maximum radius
    max_val = max(values_orig.max(), values_cpp.max())
    ax.set_rmax(max_val * 1.1)

    # Define and show radial (y) ticks
    r_ticks = np.linspace(0, max_val, 5)
    ax.set_rticks(r_ticks)
    ax.tick_params(axis='both', labelsize=15)  # Set tick font size to match caption size
    ax.set_rlabel_position(0)  # angles are in degrees; 0 means "top" or "north"

    # Set category labels
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(dimensions, fontsize=15)  # Match caption font size

    # Add legend
    ax.legend(loc="upper right", bbox_to_anchor=(1.3, 1.1), fontsize=15)
    
    # Show the plot
    plt.show()

cpp_law_school_race

# List of datasets
names = [
    "Bank (Age)", "COMPAS (Race)", "COMPAS (Gender)", "German (Age)", "German (Gender)",
    "Law School GPA (Gender)", "Law School GPA (Race)", "Law (Gender)", "Law (Race)", "MEPS (Race)"
]

datasets = [
    cpp_bank, cpp_compas_race, cpp_compas_sex, cpp_german_age, cpp_german_sex,
    cpp_law_school_gender, cpp_law_school_race, cpp_law_sex, cpp_law_race, cpp_meps19
]

find_max_gap_and_plot(datasets, names)
```
