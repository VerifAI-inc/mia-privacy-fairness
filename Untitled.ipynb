```python
import pandas as pd
import ast

```

```python
# Define feature names
feature_names = ["decile3", "decile1", "race", "cluster", "lsat", "ugpa", "dob_yr", "fulltime", "fam_inc", "gender", "tier", "indxgrp", "indxgrp2"]

# Read the CSV file
file_path = "./dp_mia2_results/rf_mia2_law_sex_shap_values_2025-02-24_eps1.csv"  # Update this with the actual path

# Load CSV, ensuring proper parsing of list values
df = pd.read_csv(file_path, index_col=0, converters={col: ast.literal_eval for col in ["orig", "transf", "dir", "reweigh", "eg"]})

df
```

```python

# Initialize a dictionary to store the results
results = {"Feature Name": feature_names}

# Process each setup separately
setups = ["orig", "transf", "dir", "reweigh", "eg"]

for setup in setups:
    shap_values = df[setup].tolist()  # Extract list of lists
    
    # Convert into a DataFrame where rows are runs and columns are features
    shap_matrix = pd.DataFrame(shap_values, columns=feature_names)
    
    # Compute mean and std for each feature
    results[f"{setup}_mean"] = shap_matrix.mean().values
    results[f"{setup}_std"] = shap_matrix.std().values

# Convert results to DataFrame
final_df = pd.DataFrame(results)

final_df
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Visualization: Feature importance based on SHAP values
plt.figure(figsize=(12, 6))

# Compute average SHAP importance across all setups
shap_importance = final_df[[col for col in final_df.columns if "_mean" in col]].mean(axis=1)

# Sort features by importance
sorted_idx = np.argsort(shap_importance)[::-1]

# Plot
plt.barh(np.array(feature_names)[sorted_idx], shap_importance[sorted_idx], color='skyblue')
plt.xlabel("Mean Absolute SHAP Value")
plt.ylabel("Feature")
plt.title("Feature Importance based on SHAP Values")
plt.gca().invert_yaxis()  # Highest importance at the top
plt.show()
```
