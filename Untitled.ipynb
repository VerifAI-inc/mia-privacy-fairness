```python
import pandas as pd
import matplotlib.pyplot as plt
import glob

# Step 1: Read all CSV files (one per dataset)
# Change the pattern if your files are located elsewhere.
csv_files = glob.glob("new_results/lira_train_test_accuracies/*network_train_test_accuracies.csv")  # all CSV files in the current directory
```

```python
csv_files
```

```python
# These will store the dataset names and computed gap values for each method.
dataset_order = []
summary_data = {}

# Methods we are interested in (we ignore syn_target)
methods = ["orig", "syn", "dir", "rew", "egr"]

for file in csv_files:
    # Use the filename (without extension) as the dataset name
    dataset_name = file.split('.')[0].split('/')[-1].split('_neural')[0].split('lira_')[-1]
    dataset_order.append(dataset_name)
    

    
    # Read CSV file
    df = pd.read_csv(file)
    

    # Step 2: Filter rows for subpopulations.
    # We use rows starting with "Train:" and "Test:".
    train_df = df[df["Accuracy Metric"].str.startswith("Train:")].copy()
    test_df = df[df["Accuracy Metric"].str.startswith("Test:")].copy()

    
    # Extract the subpopulation identifier (everything after the colon)
    train_df["subpop"] = train_df["Accuracy Metric"].apply(lambda x: x.split(":", 1)[1].strip())
    test_df["subpop"] = test_df["Accuracy Metric"].apply(lambda x: x.split(":", 1)[1].strip())
    
    # Merge the train and test data on subpopulation names
    merged = pd.merge(train_df, test_df, on="subpop", suffixes=("_train", "_test"))
    


    # Step 3: Compute the gap (train - test) for each method and average over subpopulations.
    gap_means = {}
    for m in methods:
        # Convert values to numeric (in case they are strings)
        merged[m + "_train"] = pd.to_numeric(merged[m + "_train"], errors="coerce")
        merged[m + "_test"] = pd.to_numeric(merged[m + "_test"], errors="coerce")
        gap = merged[m + "_train"] - merged[m + "_test"]
        gap_means[m] = gap.mean()
    
    summary_data[dataset_name] = gap_means

# Create a summary DataFrame with datasets as rows and methods as columns.
summary_df = pd.DataFrame(summary_data).T
summary_df.columns = ["Mean Gap (Original)", "Mean Gap (SYN)", "Mean Gap (DIR)", "Mean Gap (REW)", "Mean Gap (EGR)"]

dataset_order_1 = ['bank',
 'law_race',
 'german_sex']

dataset_labels = dataset_order_1
```

```python
# Define horizontal offsets for each mitigation method (relative to the original)
offsets = {"syn": -0.2, "dir": -0.1, "rew": 0.1, "egr": 0.2}

plt.figure(figsize=(12, 6))
for i, dataset in enumerate(dataset_order_1):
    # Get the mean gap values for this dataset
    orig = summary_df.loc[dataset, "Mean Gap (Original)"]
    
    # Plot the original point at x coordinate i.
    plt.scatter(i, orig, color="blue", label="Original" if i == 0 else "", s=100)
    
    # For each mitigation method, use the offset so the points appear near the original.
    for method, offset in offsets.items():
        if method == "syn":
            val = summary_df.loc[dataset, "Mean Gap (SYN)"]
            color = "green"
        elif method == "dir":
            val = summary_df.loc[dataset, "Mean Gap (DIR)"]
            color = "red"
        elif method == "rew":
            val = summary_df.loc[dataset, "Mean Gap (REW)"]
            color = "orange"
        elif method == "egr":
            val = summary_df.loc[dataset, "Mean Gap (EGR)"]
            color = "purple"
        
        # Plot the mitigation method point with a slight horizontal shift.
        plt.scatter(i + offset, val, color=color, label=method.upper() if i == 0 else "", s=100)
        # Draw an arrow from the original point to the mitigation point.
        plt.arrow(i, orig, offset, val - orig, color=color, head_length=0.005, length_includes_head=True)

# Customize the plot labels and legend.
plt.xticks(range(len(dataset_order_1)), dataset_labels, rotation=45, fontsize=13)
plt.ylabel("Mean Train/Test Gap", fontsize=15)
plt.legend(fontsize=13)
plt.grid(alpha=0.5)
plt.tight_layout()
plt.show()
```
